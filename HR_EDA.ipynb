# Tải thư viện và nhập dữ liệu

```python
# Nhập thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tải dữ liệu
file_path = r'D:\Download\HR Employee data.csv\HR Employee data.csv'
df = pd.read_csv(file_path)
```

# Bước 2: Kiểm tra dữ liệu

Kiểm tra giá trị thiếu

```python
# Tính số lượng và tỷ lệ phần trăm giá trị thiếu
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

# Kết hợp số lượng và phần trăm giá trị thiếu thành DataFrame
missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage
}).sort_values(by='Missing Values', ascending=False)

# In ra danh sách các cột có giá trị thiếu
print(missing_summary)
```

```python
# Xóa cột
miss_columns_to_drop = ['LeavingYear', 'Reason', 'RelievingStatus']
df = df.drop(columns=miss_columns_to_drop, axis=1)
```

```python
# Xem tổng số giá trị bị trùng
print(df.duplicated().sum())
```

# Bước 3: Trực quan dữ liệu

Ma trận dữ liệu dạng số

```python
# Phân phối của các cột số
num_columns = ['Age','DailyRate', 'DistanceFromHome', 'HourlyRate','MonthlyIncome', 'MonthlyRate',
               'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears', 'YearsAtCompany']
for col in num_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Phân phối của {col}')
    plt.xlabel(col)
    plt.ylabel('Tần suất')
    plt.show()
```

Biểu đồ tương quan giữa các biến số

Mô tả dữ liệu phân loại

```python
# Xem số lượng cột phân loại
cat_columns = ['BusinessTravel', 'Department', 'EducationField', 'Gender',
                'MaritalStatus', 'OverTime','office_code','JobLevel_updated']
# Biểu đồ cột cho các cột phân loại
for col in cat_columns:
    plt.figure(figsize=(9, 5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f"{col}")
    plt.xticks()
    plt.show()
```

```python
# Phân phối cột mục tiêu
print(df['Attrition'].value_counts())
# Show pie plot
df['Attrition'].value_counts().plot.pie(autopct='%.2f')
```

Mối quan hệ giữa các biến số, biến phân loại và biến mục tiêu Attrition

```python
# Vẽ boxplot cho từng biến
for col in num_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Attrition', y=col, palette='coolwarm')
    plt.title(f'Mối quan hệ giữa {col} và Attrition')
    plt.xlabel('Attrition')
    plt.ylabel(col)
    plt.show()
```

```python
# Vẽ countplot cho từng biến
for col in cat_columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x=col, hue='Attrition', palette='viridis')
    plt.title(f'Mối quan hệ giữa {col} và Attrition')
    plt.xlabel(col)
    plt.ylabel('Tần suất')
    plt.xticks()
    plt.legend(title='Attrition', loc='upper right')
    plt.show()
```

Mã hóa dữ liệu

```python
# Dữ liệu gồm 2 giá trị
attrition_mapping = {'Yes': 1, 'No': 0}
gender_mapping = {'Female': 1, 'Male': 0}
overtime_mapping = {'Yes': 1, 'No': 0}

df['Gender'] = df['Gender'].map(gender_mapping)
df['OverTime'] = df['OverTime'].map(overtime_mapping)
df['Attrition'] = df['Attrition'].map(attrition_mapping)

# Dữ liệu còn lại
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
categorical_cols = [
    'BusinessTravel','MaritalStatus', 'Department', 'EducationField']

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```
